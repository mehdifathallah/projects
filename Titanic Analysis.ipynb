```python
def male_female_child(passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []
for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['cabin']  
sns.factorplot('cabin',data=cabin_df,palette='winter_d',kind='count',order=list('ABCDEFG'))
```

```python
cabin_df = cabin_df[cabin_df.cabin != 'T']
sns.factorplot('cabin',data=cabin_df,palette='summer',kind='count',order=list('ABCDEFG'))
```

```python
sns.factorplot('Embarked',data=titanic_df,hue='Pclass',order=list('CQS'),palette='summer',kind='count')
```
